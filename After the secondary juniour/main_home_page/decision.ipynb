# STUDENTS AFTER CLASS 10TH GUIDANCE


###  - THIS WILL PROVIDE THE GUIDANCE FOR THE STEPS STUDENTS CAN TAKE AFTER THE CLASS 10 OR EVEN IN BEFORE 

```python
#stream majority of the student choose after the class 10th accorind to the major sources 
import pandas as pd
data = pd.read_csv('stream_selection_data.csv')
df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt
lables = df['Stream / Education Choice']
distribution = df['Regret Rate within 3 Years']
plt.style.use('ggplot')
plt.pie(distribution ,
       labels = lables,
       shadow = True,
       autopct = '%1.1f%%',wedgeprops = {'edgecolor':'black','linewidth':2,'linestyle':'--'})
plt.title('Regret Rate After 3 year with certain stream')
plt.show()
```
