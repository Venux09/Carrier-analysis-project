```python
#data representaion for the how the awared student carrier growth will 
import  pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 

data_compare = pd.read_csv('comparision_normal_vs_extraordinary.csv')
data_compare

#dividing both the data for the comparsion 

student_a = data_compare.loc[:9]
student_b = data_compare.loc[10:19]


# data representaion 
#for student a 
student_a_data = student_a['Path_Choice_Score']
student_a_success_score = student_a['Success_Score']
# for student b
student_b_data = student_b['Path_Choice_Score']
student_b_success_score = student_b['Success_Score']

#plotting the data 
plt.scatter(student_a_data,student_a_success_score,label = "student A", color = 'blue')
plt.scatter(student_b_data,student_b_success_score,label = "student B", color = 'black')

plt.title('Student A vs Student B')
plt.xlabel('CHOICE MAKED')
plt.ylabel('Result GOT ')
plt.tight_layout()
plt.legend()
plt.grid(True)
plt.show()
```
